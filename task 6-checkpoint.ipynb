```python
#lets find the answer of first question.
#1.  How many titles are there in each genre? 
import pandas as pd

# Load the data
disney_data = pd.read_csv('C:\\Users\\murar\\Downloads\\disney_plus_titles.csv')

# Count the number of titles in each genre category (from the 'listed_in' column)
# Splitting by comma since each title can belong to multiple genres
genre_counts = disney_data['listed_in'].str.split(', ').explode().value_counts()
print(genre_counts)
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart of genre counts
plt.figure(figsize=(10, 8))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Titles in Each Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot a pie chart of genre counts
plt.figure(figsize=(10, 10))
genre_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, legend=False)
plt.title('Distribution of Titles in Each Genre')
plt.ylabel('')  # Remove the default y-label
plt.show()
```

```python
# Code for TYPE column
# 0 represents Movie
# 1 represents TV Show

# Type column analysis
# Counting how many Movies and TV Shows are in the dataset
df['type_binary'] = df['type'].map({'Movie': 0, 'TV Show': 1})
type_counts = df['type_binary'].value_counts()
print("Type Counts (0 = Movie, 1 = TV Show):")
print(type_counts)

# Code for RATING column
# 0 represents General Audience (G, TV-G, TV-Y)
# 1 represents Parental Guidance (PG, TV-PG, TV-Y7, TV-Y7-FV)
# 2 represents Teen (PG-13, TV-14)
```

```python
import matplotlib.pyplot as plt

# Counting the values in the 'type' column
type_counts = df['type'].value_counts()
print("Type Counts:")
print(type_counts)

# Plotting a pie chart for 'type' column
type_counts.plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%', startangle=90)
plt.title('Movie vs TV Show Ratio')
plt.legend(['Movie', 'TV Show'])
plt.ylabel('')  # Hides the default ylabel
plt.show()
```

```python
# Let's find the answer of our 2nd question.
# Which titles have the highest runtime, and what are they?
# Ensure duration is in integer format (as done before)
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(int)

# Create bins for runtimes to categorize them
bins = [0, 60, 90, 120, 150, 200]  # Adjust these bins as desired
labels = ['<1hr', '1-1.5hrs', '1.5-2hrs', '2-2.5hrs', '2.5hrs+']
movies_df['runtime_category'] = pd.cut(movies_df['duration'], bins=bins, labels=labels)

# Use crosstab to see how many movies fall into each runtime category
runtime_crosstab = pd.crosstab(index=movies_df['runtime_category'], columns='count')

print(runtime_crosstab)

```

```python
# Let's move to question 3
#  What are the top 10 most common release years for Disney+ titles?

# Count the occurrences of each release year to find the most common release years
release_year_counts = df['release_year'].value_counts().head(10)

# Display the result
print(release_year_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the top 10 most common release years
top_release_years = df['release_year'].value_counts().head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_release_years.index, y=top_release_years.values, palette='viridis')
plt.title("Top 10 Most Common Release Years for Disney+ Titles")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Use crosstab to count occurrences of each release year
release_year_crosstab = pd.crosstab(index=df['release_year'], columns='count').sort_values(by='count', ascending=False).head(10)

# Display the crosstab result
print(release_year_crosstab)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame creation
# Replace this with your actual DataFrame
data = {
    'title': ['Title1', 'Title2', 'Title3', 'Title4', 'Title5', 
              'Title6', 'Title7', 'Title8', 'Title9', 'Title10', 
              'Title11', 'Title12', 'Title13', 'Title14', 'Title15'],
    'release_year': [2020, 2021, 2020, 2022, 2021, 
                     2020, 2019, 2022, 2020, 2021, 
                     2019, 2020, 2021, 2022, 2021]
}

df = pd.DataFrame(data)

# Counting the occurrences of each release year
release_counts = df['release_year'].value_counts()

# Getting the top 10 most common release years
top_10_years = release_counts.nlargest(10)

# Plotting
plt.figure(figsize=(10, 6))
top_10_years.plot(kind='bar', color='lightskyblue')
plt.title('Top 10 Most Common Release Years for Disney+ Titles')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.show()

```

```python
#Now question 4!
#4. what is the distribution between movies and TV shows?
import pandas as pd

# Sample DataFrame
data = {
    'title': ['Title1', 'Title2', 'Title3', 'Title4', 'Title5'],
    'type': ['movie', 'tv show', 'movie', 'tv show', 'movie'],
}

df = pd.DataFrame(data)

# Creating a crosstab for the distribution between movies and TV shows
distribution = pd.crosstab(df['type'], columns='count')

# Display the crosstab
print(distribution)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame creation
# Replace this with your actual DataFrame
data = {
    'title': ['Title1', 'Title2', 'Title3', 'Title4', 'Title5', 
              'Title6', 'Title7', 'Title8', 'Title9', 'Title10'],
    'type': ['movie', 'tv show', 'movie', 'tv show', 'movie', 
             'tv show', 'movie', 'tv show', 'movie', 'tv show'],
}

df = pd.DataFrame(data)

# Count plot for the distribution between movies and TV shows
sns.countplot(x='type', data=df)

# Adding title and labels
plt.title('Distribution Between Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')

# Show the plot
plt.show()
```
